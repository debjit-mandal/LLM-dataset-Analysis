```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Large language models (2024).csv'
df = pd.read_csv(file_path, encoding='latin1')

df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]

df['parameters'] = pd.to_numeric(df['parameters'], errors='coerce')
df['tokens'] = pd.to_numeric(df['tokens'], errors='coerce')
df['alscore'] = (df['parameters'] * df['tokens']) ** 0.5

df.head()
```


# Model Count by Company

This section shows the count of language models developed by each company.


```python

# Model count by company
model_count_by_company = df['comapany'].value_counts()

# Plotting model count by company with fixed x-axis labels
plt.figure(figsize=(15, 8))
sns.barplot(x=model_count_by_company.index, y=model_count_by_company.values)
plt.title('Count of Models by Company')
plt.xlabel('Company')
plt.ylabel('Number of Models')
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```
