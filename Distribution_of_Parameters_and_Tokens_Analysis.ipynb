```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Large language models (2024).csv'
df = pd.read_csv(file_path, encoding='latin1')

df.columns = [col.strip().replace(' ', '_').lower() for col in df.columns]

df['parameters'] = pd.to_numeric(df['parameters'], errors='coerce')
df['tokens'] = pd.to_numeric(df['tokens'], errors='coerce')
df['alscore'] = (df['parameters'] * df['tokens']) ** 0.5

df.head()
```


# Distribution of Parameters and Tokens

This section shows the distribution of the number of parameters and tokens across different language models.


```python

# Plotting the distribution of parameters and tokens
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['parameters'].dropna(), kde=True, bins=20)
plt.title('Distribution of Parameters (in Billions)')
plt.xlabel('Parameters (Billions)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['tokens'].dropna(), kde=True, bins=20)
plt.title('Distribution of Tokens (in Billions)')
plt.xlabel('Tokens (Billions)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
